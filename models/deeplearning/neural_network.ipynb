**DEEP LEARNING**

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Input,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
```

**UPLOAD DATASET**

```python
df = pd.read_csv("../../dataset/hearts_stats.csv") # Import dataset
df.head(10)
```

```python
df.info()
```

```python
df.describe().transpose()
```

**DATA CLEARING**

```python
print(f"Null values: {df.isnull().sum()}\n\n") # Check null values
print(f"Na values: {df.isna().sum()}\n\n") # Check na values
print(f"Duplicates: {df.duplicated().sum()}") # Check duplicated
```

```python
df = df.drop_duplicates() # Drop duplicates
df.head()
```

```python
df.info()
```

```python
print(f"Duplicates: {df.duplicated().sum()}") # Check duplicated
```

**SPLIT INTO LABELS AND TARGETS**

```python
X = df.drop("target",axis=1)
y = df["target"]
```

**MAKE TRAIN AND TEST DATA**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=911)
```

**ADD SCALER FOR UPGRADE MODEL**

```python
scaler = MinMaxScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

**CHECK SHAPE OF DATA**

```python
scaled_X_train.shape
```

```python
scaled_X_test.shape
```

**CREATE NEURAL NETWORK**

```python
model = Sequential()

model.add(Input(shape=(11,)))
model.add(Dense(256, kernel_regularizer=tf.keras.regularizers.l2(0.002), activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(128, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(64, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(32, activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.4))

model.add(Dense(1, activation="sigmoid"))

optimizer = Adam(learning_rate=0.001)

model.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])
```

**PARAMS AND DESCRIBE**

```python
model.summary()
```

**FIT NEURAL NETWORK**

```python
model.fit(scaled_X_train,y_train,epochs=1000,callbacks=[EarlyStopping(patience=14)],validation_data=(scaled_X_test,y_test))
```

**MAKE DATAFRAME ABOUT LOSS AND ACCURACY**

```python
history = pd.DataFrame(model.history.history)
```

**SEE WHAT THE TRAINING PROCESS**

```python
history.plot()
```

**LOSS REPORT**

```python
plt.plot(history['loss'], label='Training Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

**ACCURACY REPORT**

```python
plt.plot(history['accuracy'], label='Training Accuracy')
plt.plot(history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()
```

```python
# For test model
def prediction_evaluate(model,name):
    pred =(model.predict(scaled_X_test) > 0.5).astype("int32")
    cm = confusion_matrix(y_test, pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[1,0])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(name)
    plt.show()
    
    print(classification_report(y_test, pred))
    print("\n\n\n\n\n")
```

**RESULT NEURAL NETWORK**

```python
prediction_evaluate(model,"ANN")
```

**ACCURACY NEURAL NETWORK: 89%-90%**

**MAKE MODEL WITH 10000 EPOCHS**

```python
model_fun = Sequential()

model_fun.add(Input(shape=(11,)))
model_fun.add(Dense(256, kernel_regularizer=tf.keras.regularizers.l2(0.1), activation="relu"))
model_fun.add(BatchNormalization())
model_fun.add(Dropout(0.4))

model_fun.add(Dense(128, activation="relu"))
model_fun.add(BatchNormalization())
model_fun.add(Dropout(0.4))

model_fun.add(Dense(64, activation="relu"))
model_fun.add(BatchNormalization())
model_fun.add(Dropout(0.4))

model_fun.add(Dense(32, activation="relu"))
model_fun.add(BatchNormalization())
model_fun.add(Dropout(0.4))

model_fun.add(Dense(1, activation="sigmoid"))

optimizer = Adam(learning_rate=0.001)

model_fun.compile(optimizer=optimizer, loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model_fun.fit(scaled_X_train,y_train,epochs=10000,validation_data=(scaled_X_test,y_test))
```

```python
history_fun = pd.DataFrame(model_fun.history.history)
```
